```python

movies=pd.read_csv('tmdb_movies_data.csv')
movies = movies.fillna("")
```

```python
movies
```

```python
movies=movies[['id','imdb_id','original_title', 'genres', 'overview', 'cast', 'director', 'keywords']]
```

```python
#movies
```

```python
movies["cast"] = movies["cast"].str.split("|")

movies["genres"] = movies["genres"].str.split("|")
movies["keywords"] = movies["keywords"].str.split("|")
movies["director"] = movies["director"].apply(lambda x: [x])

```

```python
#movies
```

```python
movies["overview"] = movies["overview"].fillna("").astype(str)
movies["overview"] = movies["overview"].apply(lambda x: x.split())
```

```python
#movies
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['director']=movies['director'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])


movies["cast"] = movies["cast"].apply(lambda x: x[:3] if isinstance(x, list) else x)

```

```python
#movies
```

creating the tags column 

```python
movies['tags']= movies['overview'] +movies['genres'] +movies['cast']+movies['director']+movies['keywords']
```

```python
movies
```

creating new dataframe for "title","movie_id"and"Tags" column 

```python
df = movies[['id','imdb_id', 'original_title','tags']]
```

```python
df
```

convertung the list data of column into a string 

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower())
```

```python
df
```

finally steming the repeated words

```python
import nltk 
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
df['tags']=df['tags'].apply(stem)
```

```python
# df.info()
```

here all the data is in its final state 

NOW TO START THE VECTORIZATION 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000, stop_words='english')

vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
# cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index=df[df['original_title']==movie].index[0]
    proper_index=movie_index.item()

    distance= similarity[proper_index]

    movies_list=sorted(list(enumerate(distance)),reverse=True,  key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(df.iloc[i[0]].original_title)
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
